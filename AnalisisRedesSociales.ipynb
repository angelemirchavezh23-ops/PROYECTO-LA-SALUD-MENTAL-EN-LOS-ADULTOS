## Identificar el Rango de Edad con Más Compras

### Subtask:
Analiza la columna `Rango_Edad` en relación con `Compra_Tras_Anuncio` para determinar qué grupo de edad ha realizado más compras después de ver anuncios en redes sociales.


**Reasoning**:
To determine the age range with the most purchases after seeing social media ads, I will first filter the `df_redesSociales` DataFrame to include only those rows where `Compra_Tras_Anuncio` is 'Sim'. Then, I will count the occurrences of each unique value in the `Rango_Edad` column from this filtered data to get the distribution of purchases by age range. Finally, I will identify the age range with the highest number of purchases.

**Reasoning**:
To identify the age range with the most purchases after seeing social media ads, I will filter the DataFrame for 'Sim' purchases, calculate the frequency distribution of age ranges, and then extract the top age range.



```python
compras_tras_anuncio = df_redesSociales[df_redesSociales['Compra_Tras_Anuncio'] == 'Sim']

compras_por_rango_edad = compras_tras_anuncio['Rango_Edad'].value_counts()

print("Distribución de compras tras anuncio por rango de edad:")
print(compras_por_rango_edad)

rango_edad_mas_compras = compras_por_rango_edad.index[0]
print(f"\nEl rango de edad con más compras después de ver anuncios en redes sociales es: {rango_edad_mas_compras}")
```

## Determinar la Red Social Más Popular

### Subtask:
Procesa la columna `Redes_Sociales_Usadas` para contar la frecuencia de cada plataforma y encontrar cuál es la más utilizada.


**Reasoning**:
To determine the most popular social media platform, I will first split the 'Redes_Sociales_Usadas' column by the semicolon delimiter, then explode the resulting lists into individual rows, count the occurrences of each platform, and finally identify the platform with the highest frequency.



```python
social_media_platforms = df_redesSociales['Redes_Sociales_Usadas'].str.split(';').explode()

# Count the frequency of each platform
frecuencia_plataformas = social_media_platforms.value_counts()

print("Frecuencia de cada plataforma de redes sociales:")
print(frecuencia_plataformas)

# Identify the most popular platform
plataforma_mas_popular = frecuencia_plataformas.index[0]
print(f"\nLa plataforma de redes sociales más popular es: {plataforma_mas_popular}")
```
